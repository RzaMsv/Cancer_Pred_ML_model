```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
# Libraries for logistic regression model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

## Initial inspection of the data

```python
df = pd.read_csv("BreastCancer.csv")
df.head()
```

## Dropping first and last column

```python
df.drop('id' ,axis=1 ,inplace=True)
df.drop('Unnamed: 32' ,axis=1 ,inplace=True)
df.head()
```

## Now let's change M and B in diagnosis col to 1 and 0

```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
df.head()
```

## Comparing the number of Malignant to Benign patients

```python
plt.hist(df['diagnosis'])
plt.title('Diagnosis (M=1 , B=0)')
plt.show()
```

## Now we need to select feature with highest explained variances

```python
df.describe()
```

```python
y = df.iloc[:, 0]
X = df.iloc[:, 1:31]
X.head()
```

## Before everything, let's generate a heatmap between features pairwise correlation

```python
corr_matt = X.corr(method='pearson')
plt.figure(figsize=(20,20))
corr = sns.heatmap(corr_matt, annot=True, cmap='Greens', cbar=False)
```

## Now we split our data to train and test datasets
### We use 70/30 ratio

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create the Logistic Regression model
model = LogisticRegression()
# Fit the model to the training data
model.fit(X_train, y_train)
```

## Make predictions on the test data

```python
y_pred = model.predict(X_test)
```

## Now we evaluate our model

```python
# To calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# To display other classification metrics
print(classification_report(y_test, y_pred))

# to display the confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

## Here, we calculated the accuracy of cancer prediction with an acuracy of 97%
